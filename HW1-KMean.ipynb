```python
import math
import numpy as np
from matplotlib import pyplot as plt
def cal_distance(x,y):
  return math.sqrt((x[0]-y[0])**2+(x[1]-y[1])**2)
def find_cluster(pos):
  dis=1e9
  clus=0
  i=0
  for c in centroid:
    if dis> cal_distance(c,pos):
      dis=cal_distance(c,pos)
      clus=i
    i+=1
  return clus
def is_equal(a,b):
  return abs(a[0]-b[0])<0.001 and abs(a[1]-b[1])<0.001
def cal_cluster():
  global centroid,position
  change = True
  while change:
    change = False
    clusters = [[] for _ in range(3)]
    for pos in position:
      clusters[find_cluster(pos)].append(pos)
    c1 = np.mean(clusters[0],axis=0)
    c2 = np.mean(clusters[1],axis=0)
    c3 = np.mean(clusters[2],axis=0)
    if is_equal(c1, centroid[0])==False or is_equal(c2, centroid[1])==False or is_equal(c3, centroid[2])==False:
      change = True
    centroid[0]=c1
    centroid[1]=c2
    centroid[2]=c3
  print(centroid)
  return clusters
print("Centroid:T5")
centroid =np.array([[3,3],[2,2],[-3,-3]],dtype=float)
position =np.array([[1,2],[3,3],[2,2],[8,8],[6,6],[7,7],[-3,-3],[-2,-4],[-7,-7]],dtype=float)
cluster = cal_cluster()
x =[]
y=[]
for i in cluster[0]:
  x.append(i[0])
  y.append(i[1])
plt.scatter(x,y,color='red')
x =[]
y=[]
for i in cluster[1]:
  x.append(i[0])
  y.append(i[1])
plt.scatter(x,y,color='blue')
x =[]
y=[]
for i in cluster[2]:
  x.append(i[0])
  y.append(i[1])
plt.scatter(x,y,color='green')
plt.show()
```
